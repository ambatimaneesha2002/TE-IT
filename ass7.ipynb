```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data from the CSV file
data = pd.read_csv("newfb.csv")
print(data)
```

```python
# Create a bar plot of Total Interactions for each post type
plt.figure(figsize=(12, 6))
sns.barplot(x='Type', y='Total Interactions', data=data)
plt.title('Total Interactions by Post Type')
plt.xlabel('Post Type')
plt.ylabel('Total Interactions')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the number of occurrences of each type of post
post_counts = df['Type'].value_counts()

# Create a pie chart
plt.pie(post_counts, labels=post_counts.index, autopct='%1.1f%%')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('fb.csv')

# Create a boxplot of Lifetime Post Total Reach by Type
sns.boxplot(x='Type', y='Lifetime Post Total Reach', data=df)

# Add labels and title
plt.xlabel('Post Type')
plt.ylabel('Lifetime Post Total Reach')
plt.title('Boxplot of Lifetime Post Total Reach by Post Type')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('fb.csv')

# Create a countplot of Type
sns.countplot(x='Type', data=df)

# Add labels and title
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Countplot of Type')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('fb.csv')

# Create a scatterplot of Lifetime Post Total Reach vs Page Total Likes
plt.scatter(df['Page total likes'], df['Lifetime Post Total Reach'])

# Add labels and title
plt.xlabel('Page Total Likes')
plt.ylabel('Lifetime Post Total Reach')
plt.title('Scatterplot of Lifetime Post Total Reach vs Page Total Likes')

# Show the plot
plt.show()
```
